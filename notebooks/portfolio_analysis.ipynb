```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Data

```python
df = pd.read_csv("portfolio_data.csv", parse_dates=['date'])
print(df.head())
```

NAV over time for all verticals

```python
pivot_nav = df.pivot(index='date', columns='vertical', values='nav')
pivot_nav.plot(figsize=(12, 6), title='NAV Over Time')
plt.ylabel('NAV')
plt.grid(True)
plt.show()
```

Average daily return per vertical

```python
avg_returns = df.groupby('vertical')['daily_return'].mean().sort_values(ascending=False)
sns.barplot(x=avg_returns.index, y=avg_returns.values)
plt.title("Average Daily Return by Vertical")
plt.ylabel("Daily Return")
plt.xticks(rotation=45)
plt.show()
```

Allocation heatmap over time

```python
pivot_alloc = df.pivot(index='date', columns='vertical', values='allocation_percentage')
sns.heatmap(pivot_alloc.T, cmap="YlGnBu", cbar_kws={'label': 'Allocation %'})
plt.title("Daily Allocation % by Vertical")
plt.xlabel("Date")
plt.ylabel("Vertical")
plt.show()
```
